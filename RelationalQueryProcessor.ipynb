```python
from sqlite3 import connect
from pandas import read_csv, Series, DataFrame, read_sql
import pandas as pd
pd.options.mode.chained_assignment = None


def getAllAnnotations():
    with connect("relational.db") as con: 
        query_1 = "SELECT annotation,target,body,motivation FROM annotationsId LEFT JOIN imageId ON annotationsId.imageId == imageId.imageId"
        all_annotations_query = read_sql(query_1,con)
    return all_annotations_query
getAllAnnotations()

#It's strange isn't it, why do I have to combine two tables to know something that is already there in the raw data? 
#An annotation should be composed with a target and a body, I don't think they should be separated in two tables.
#What about uploading the annotataion raw data into the database? It could be much easier in this way.
```
